```python
freq = 0
actline = 0
freqArr = dict()
```

```python
for i in range(200):
    data = open('data/day1.txt','r')
    for line in data:
      value = line.replace('\n','')
      actline += 1
      freq += int(value)
      #print(actline, freq)
      try:
        freqArr[freq] += 1
        print('freq found', actline, freq)
      except:
        freqArr[freq] = 1
    
    data.close()
```
