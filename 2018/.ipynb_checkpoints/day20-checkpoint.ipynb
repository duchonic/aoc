````python
import networkx as nx

def solveMaze(data):

    X = 0
    Y = 1
    G = nx.DiGraph()

    position = [0,0]
    oldpos = (0,0)
    cross = list()

    for direction in data[1:-1]:
        if direction == 'E':
            position[X] += 1
        elif direction == 'W':
            position[X] -= 1
        elif direction == 'N':
            position[Y] += 1
        elif direction == 'S':
            position[Y] -= 1
        elif direction == '(':
            cross.append(pos)
            #print('cross append:', pos)
        elif direction == '|':
            if len(cross) > 0:
                oldpos = cross.pop()
                position = [oldpos[X],oldpos[Y]]
                #print('cross pop:', oldpos)

        pos = (position[X],position[Y])
        G.add_node(pos)
        G.add_edge(oldpos, pos)
        oldpos = pos

    nx.draw(G, with_labels=True)    
    
    
    return 1
````
