```python
BANR = 0
EQRR = 1
SETR = 2
EQIR = 3
BORI = 4
MULI = 5
BANI = 6
BORR = 7
GTIR = 8
GTRR = 9
ADDI = 10
GTRI = 11
EQRI = 12
ADDR = 13
MULR = 14
SETI = 15
```

```python
#opCode
#OPCODE = 0
A      = 1
B      = 2
C      = 3

def calc(inReg, opCode):
    
    #print(inReg)
    #print(outRegShould)
    code = opCode[0]
    
    outReg = list(inReg)

    if code == ADDR:
        #addr
        outReg[opCode[C]] = inReg[opCode[A]] + inReg[opCode[B]]
    elif code == ADDI:
        #addi
        outReg[opCode[C]] = inReg[opCode[A]] + opCode[B]
    elif code == MULR:
        #mulr
        outReg[opCode[C]] = inReg[opCode[A]] * inReg[opCode[B]]
    elif code == MULI:
        #muli
        outReg[opCode[C]] = inReg[opCode[A]] * opCode[B]
    elif code == BANR:
        #banr bitwise and
        outReg[opCode[C]] = inReg[opCode[A]] & inReg[opCode[B]]
    elif code == BANI:
        #bani bitwise and
        outReg[opCode[C]] = inReg[opCode[A]] & opCode[B]
    elif code == BORR:
        #borr bitwise or
        outReg[opCode[C]] = inReg[opCode[A]] | inReg[opCode[B]]
    elif code == BORI:
        #bori bitwise or
        outReg[opCode[C]] = inReg[opCode[A]] | opCode[B]
    elif code == SETR:
        #setr
        outReg[opCode[C]] = inReg[opCode[A]]
    elif code == SETI:
        #seti
        outReg[opCode[C]] = opCode[A]
    elif code == GTIR:
        #gtir
        outReg[opCode[C]] = opCode[A] > inReg[opCode[B]]
    elif code == GTRI:
        #gtri
        outReg[opCode[C]] = inReg[opCode[A]] > opCode[B]
    elif code == GTRR:
        #gtrr
        outReg[opCode[C]] = inReg[opCode[A]] > inReg[opCode[B]]
    elif code == EQIR:
        #eqir
        outReg[opCode[C]] = opCode[A] == inReg[opCode[B]]
    elif code == EQRI:
        #eqri
        outReg[opCode[C]] = inReg[opCode[A]] == opCode[B]
    elif code == EQRR:
        #eqrr
        outReg[opCode[C]] = inReg[opCode[A]] == inReg[opCode[B]]
    else:
        assert 0
        
    print(inReg, code, opCode, outReg)
    return outReg
```

```python
start = [1, 0, 2, 1]
data = open('data/day16_2.data','r')
```

```python
for line in data:
    codeStr = line
    code = codeStr.split()    
    
    for i in range(4):
        code[i] = int(code[i])
    newStart = calc(start, code)
    start = newStart
    #print('start:',start, 'newStart:', newStart)
```
