````python
class Object():
    patchNr = 1
    
    def __init__(self, pos, velocity):
        self.x = pos[0]-100
        self.y = pos[1]-147
        self.speedx = velocity[0]
        self.speedy = velocity[1]
        self.patchNr = self.__class__.patchNr
        self.__class__.patchNr += 1
        
    def __str__(self):
        return f'i am patch nr {self.patchNr}'
            
    def move(self, sec):
        self.x += self.speedx*sec
        self.y += self.speedy*sec
        
    def pos(self):
        return [self.x, self.y]
        
    def getInfo(self):
        return {'objectnr':self.patchNr}
````

````python
class Screen():
    patch = 1
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def show(self, patchObjects):
        
        board = []
        
        for row in range(self.y):
            board.append([])
            for col in range(self.x):
                board[row].append('.')
        
        for item in patchObjects:
            pos = item.pos()
            board[pos[1]][pos[0]] = '#' #item.getInfo()['objectnr']
         
        
        for lines in board:
            print("".join(str(x) for x in lines) )       
        
        return board
````

````python
#data = open('data/day10test.data','r')
data = open('data/day10.data','r')

patchObjects = []

for line in data:
    objectData = line.replace('position=<','').replace(',','').replace('>','').replace('velocity=<','').split() 
    patchObjects.append( Object( (int(objectData[0]), int(objectData[1])) , (int(objectData[2]), int(objectData[3]) ) ))
````

````python
for object in patchObjects:
    object.move(10011)
````

````python
myScreen = Screen(80, 15)
board = myScreen.show(patchObjects)
````
