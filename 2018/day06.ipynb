````python
class Object():
    patchNr = 1
    
    def __init__(self, pos):
        self.x = pos[0]
        self.y = pos[1]
        self.patchNr = self.__class__.patchNr
        self.__class__.patchNr += 1
        
    def __str__(self):
        return f'i am patch nr {self.patchNr}'
            
    def pos(self):
        return [self.x, self.y]
        
    def getInfo(self):
        return {'objectnr':self.patchNr}
````

````python
class Screen():
    patch = 1
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def show(self, patchObjects):
        
        board = []
        for y in range(self.y):
            board.append([])
            for x in range(self.x):
                board[y].append(0)
        
        
        for item in patchObjects:
            pos = item.pos()
            board[pos[1]][pos[0]] = item.getInfo()['objectnr']
                              
        def calcManhattanDistance(pos, obj):           
            return sum( (abs(pos[0] - obj.pos()[0]), abs( pos[1] - obj.pos()[1])) )
            
        def getMinDistance(pos):
            distance = []
            for item in patchObjects:
                distance.append(calcManhattanDistance(pos , item))

            if distance.count(min(distance)) > 1:
                return 0
            else:
                return distance.index(min(distance))+1
            
        calc = dict()
        for blub in range(51):
            calc.update({blub:0})
            
        for y in range(self.y):
            for x in range(self.x):
                
                minDis =  getMinDistance( (x,y) )
                
                if (y == 0) or (x == 0) or (y ==  (self.y-1) ) or (x == (self.x-1) ):
                    calc[minDis] -= 100000
                else:
                    calc[minDis] += 1
                    
                board[y][x] = minDis
                        
        return calc
````

````python
patchObjects = []
data = open('data/day6.dat', 'r')

for line in data:
    objectData = line.replace(',','').split()
    patchObjects.append( Object( (int(objectData[0]), int(objectData[1])) ))

sizex = 350
sizey = 360
    
myScreen = Screen(sizex, sizey)
board1 = myScreen.show(patchObjects)
````
