```python
#OPCODE = 0
A      = 1
B      = 2
C      = 3

def calc(inReg, opCode, outRegShould):
    
    #print(inReg)
    print(opCode)
    #print(outRegShould)
    
    total = 0
    
    outRegIs = list(inReg)
    #addr
    outRegIs[opCode[C]] = inReg[opCode[A]] + inReg[opCode[B]]
    if outRegShould == outRegIs:
        total += 1
        print('addr', outRegShould, outRegIs)
    
    outRegIs = list(inReg)
    #addi
    outRegIs[opCode[C]] = inReg[opCode[A]] + opCode[B]
    if outRegShould == outRegIs:
        total += 1
        print('addi', outRegShould, outRegIs )

    outRegIs = list(inReg)
    #mulr
    outRegIs[opCode[C]] = inReg[opCode[A]] * inReg[opCode[B]]
    if outRegShould == outRegIs:
        total += 1
        print('mulr', outRegShould, outRegIs)
        
    outRegIs = list(inReg)
    #muli
    outRegIs[opCode[C]] = inReg[opCode[A]] * opCode[B]
    if outRegShould == outRegIs:
        total += 1
        print('muli', outRegShould, outRegIs )

    outRegIs = list(inReg)
    #banr bitwise and
    outRegIs[opCode[C]] = inReg[opCode[A]] & inReg[opCode[B]]
    if outRegShould == outRegIs:
        total += 1
        print('banr', outRegShould, outRegIs )

    outRegIs = list(inReg)
    #bani bitwise and
    outRegIs[opCode[C]] = inReg[opCode[A]] & opCode[B]
    if outRegShould == outRegIs:
        total += 1
        print('bani', outRegShould, outRegIs )
        
    outRegIs = list(inReg)
    #borr bitwise or
    outRegIs[opCode[C]] = inReg[opCode[A]] | inReg[opCode[B]]
    if outRegShould == outRegIs:
        total += 1
        print('borr', outRegShould, outRegIs )

    outRegIs = list(inReg)
    #bori bitwise or
    outRegIs[opCode[C]] = inReg[opCode[A]] | opCode[B]
    if outRegShould == outRegIs:
        total += 1
        print('bori', outRegShould, outRegIs )    
        
    outRegIs = list(inReg)
    #setr
    outRegIs[opCode[C]] = inReg[opCode[A]]
    if outRegShould == outRegIs:
        total += 1
        print('setr', outRegShould, outRegIs )   
        
    outRegIs = list(inReg)
    #seti
    outRegIs[opCode[C]] = opCode[A]
    if outRegShould == outRegIs:
        total += 1
        print('seti', outRegShould, outRegIs )   
        
    outRegIs = list(inReg)
    #gtir
    outRegIs[opCode[C]] = opCode[A] > inReg[opCode[B]]
    if outRegShould == outRegIs:
        total += 1
        print('gtir', outRegShould, outRegIs )   
    
    outRegIs = list(inReg)
    #gtri
    outRegIs[opCode[C]] = inReg[opCode[A]] > opCode[B]
    if outRegShould == outRegIs:
        #total += 1
        #print('gtri', outRegShould, outRegIs )   
        pass
    
    outRegIs = list(inReg)
    #gtrr
    outRegIs[opCode[C]] = inReg[opCode[A]] > inReg[opCode[B]]
    if outRegShould == outRegIs:
        #total += 1
        #print('gtrr', outRegShould, outRegIs )
        pass
    
    outRegIs = list(inReg)
    #eqir
    outRegIs[opCode[C]] = opCode[A] == inReg[opCode[B]]
    if outRegShould == outRegIs:
        #total += 1
        #print('eqir', outRegShould, outRegIs )   
        pass
    
    outRegIs = list(inReg)
    #eqri
    outRegIs[opCode[C]] = inReg[opCode[A]] == opCode[B]
    if outRegShould == outRegIs:
        #total += 1
        #print('eqri', outRegShould, outRegIs )  
        pass
    
    outRegIs = list(inReg)
    #eqrr
    outRegIs[opCode[C]] = inReg[opCode[A]] == inReg[opCode[B]]
    if outRegShould == outRegIs:
        #total += 1
        #print('eqrr', outRegShould, outRegIs )   
        pass
        
    return total
```
